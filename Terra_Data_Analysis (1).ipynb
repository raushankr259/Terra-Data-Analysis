**2. Highest revenue with respect to neighbourhoods**

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('Terra Data Assignment.xlsx', sheet_name='Data')

# Calculate the number of booked nights
df['booked_nights'] = 365 - df['Unoccupied days\n(in last 365 days)']

# Calculate revenue for each listing
df['revenue'] = df['booked_nights'] * df['price per night']

# Group by neighborhood and calculate total revenue
neighborhood_revenue = df.groupby('neighbourhood')['revenue'].sum().reset_index()

# Sort neighborhoods by total revenue in descending order
neighborhood_revenue = neighborhood_revenue.sort_values(by='revenue', ascending=False)

# Display the neighborhoods with the highest revenue
print(neighborhood_revenue)

# Create the bar chart
fig = px.bar(neighborhood_revenue, x='neighbourhood', y='revenue',
             labels={'neighbourhood': 'Neighborhood', 'revenue': 'Total Revenue'},
             title='Highest Revenue by Neighborhood')

# Customize the appearance
fig.update_layout(xaxis_tickangle=-45,
                  xaxis_title='Neighborhood',
                  yaxis_title='Total Revenue')

fig.show()
```

**3. Best overall listings- (Weightage - 0.6 value of booking, 0.4 number of reviews)**

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel('Terra Data Assignment.xlsx', sheet_name='Data')

# Calculate the number of booked nights
df['booked_nights'] = 365 - df['Unoccupied days\n(in last 365 days)']

# Calculate a booking value (assuming price per night is a good proxy)
df['booking_value'] = df['booked_nights'] * df['price per night']

# Calculate the weighted score
df['weighted_score'] = 0.6 * df['booking_value'] + 0.4 * df['number_of_reviews']

# Sort by overall score in descending order
df_sorted = df.sort_values('weighted_score', ascending=False)

# Display the top listings
print(df_sorted[['neighbourhood', 'name', 'weighted_score']].head())

fig = px.bar(df_sorted.head(10), x='name', y='weighted_score',
             labels={'name': 'Listing Name', 'weighted_score': 'Weighted Score'},
             title='Top 10 Best Overall Listings')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

Booking Performance by Reviews (with Bucketing)


```python
import pandas as pd

# Load the Excel file
df_data = pd.read_excel('Terra Data Assignment.xlsx', sheet_name='Data')

# Calculate the number of booked nights
df_data['booked_nights'] = 365 - df_data['Unoccupied days\n(in last 365 days)']

# Define buckets for the number of reviews
bins = [0, 10, 50, 100, 200, 500, 1000]
labels = ['0-10', '11-50', '51-100', '101-200', '201-500', '501-1000']
df_data['review_bucket'] = pd.cut(df_data['number_of_reviews'], bins=bins, labels=labels, include_lowest=True)

# Group by review bucket and calculate total booked nights
booking_by_review_bucket = df_data.groupby('review_bucket')['booked_nights'].sum().reset_index()

# Sort by total booked nights in descending order
booking_by_review_bucket = booking_by_review_bucket.sort_values(by='booked_nights', ascending=False)

# Display the booking performance by review bucket
print(booking_by_review_bucket)

# Create the bar chart
fig = px.bar(booking_by_review_bucket, x='review_bucket', y='booked_nights',
             labels={'review_bucket': 'Review Bucket', 'booked_nights': 'Total Booked Nights'},
             title='Booking Performance by Review Bucket')

fig.show()

```

**Sorted list of business done by all host**

```python
import pandas as pd

# Load the Excel file
df_data = pd.read_excel('Terra Data Assignment.xlsx', sheet_name='Data')

df_data['booked_nights'] = 365 - df_data['Unoccupied days\n(in last 365 days)']

# Calculate the value of booking for each listing
df_data['value_of_booking'] = df_data['booked_nights'] * df_data['price per night']

# Group by host and calculate the total value of bookings
business_by_host = df_data.groupby('host_name')['value_of_booking'].sum().reset_index()

# Sort the hosts by total value of bookings in descending order
business_by_host = business_by_host.sort_values(by='value_of_booking', ascending=False)

# Display the sorted list of business done by all hosts
print(business_by_host.head(10))

# Create the bar chart
fig = px.bar(business_by_host.head(10), x='host_name', y='value_of_booking',
             labels={'host_name': 'Host Name', 'value_of_booking': 'Total Value of Bookings'},
             title='Top 10 Hosts by Total Value of Bookings')

# Rotate x-axis labels for better readability
fig.update_layout(xaxis_tickangle=-45)

fig.show()
```
